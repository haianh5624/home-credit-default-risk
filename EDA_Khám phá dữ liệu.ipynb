# I. Khám phá dữ liệu

### 1. Tổng quan dữ liệu gốc

```python
# Tạo một list để chứa thông tin thống kê
summary_list = []

print("=== 1. TỔNG QUAN HỆ THỐNG DỮ LIỆU  ===\n")

for key, df in dfs.items():
    # 1. Tính toán các chỉ số cơ bản
    n_rows, n_cols = df.shape
    n_dups = df.duplicated().sum()
    
    # Tính tổng số cell bị thiếu và tỷ lệ
    n_missing_cells = df.isnull().sum().sum()
    missing_ratio = (n_missing_cells / (n_rows * n_cols)) * 100
    
    # Đếm số lượng cột có chứa dữ liệu null
    cols_with_null = df.isnull().any().sum()
    
    # Tính dung lượng RAM (để biết bảng nào nặng)
    mem_usage = df.memory_usage(deep=True).sum() / 1024**2  # Convert to MB
    
    # Lưu vào list
    summary_list.append({
        'Table Name': key,
        'Rows': n_rows,
        'Cols': n_cols,
        'Duplicates': n_dups,
        'Total Missing Cells': n_missing_cells,
        'Missing (%)': round(missing_ratio, 2),
        'Null Cols Count': cols_with_null,
        'Memory (MB)': round(mem_usage, 2)
    })

    # --- IN RA MÀN HÌNH DẠNG TEXT (GIỐNG ẢNH BẠN GỬI NHƯNG ĐẸP HƠN) ---
    print(f"--- Bảng: {key} ---")
    print(f"• Kích thước   : {n_rows:,} dòng, {n_cols} cột")
    print(f"• Dòng trùng lặp: {n_dups}")
    print(f"• Dữ liệu thiếu : {n_missing_cells:,} giá trị ({round(missing_ratio, 1)}% tổng dữ liệu)")
    print(f"• Số cột có null: {cols_with_null} / {n_cols} cột")
    print("-" * 40) # Đường kẻ phân cách

# --- TẠO BẢNG TỔNG HỢP ĐỂ SO SÁNH (Cái này nhìn sẽ chuyên nghiệp hơn) ---
summary_df = pd.DataFrame(summary_list)
summary_df.set_index('Table Name', inplace=True)

print("\n\n>>> BẢNG TỔNG HỢP SO SÁNH:")
display(summary_df.style.background_gradient(cmap='Reds', subset=['Missing (%)', 'Duplicates']))
```

### 6. Khám phá về lịch sử tín dụng

```python
# Lấy dữ liệu cần thiết
df_train = dfs['app_train'][['SK_ID_CURR', 'TARGET']].copy()
bureau = dfs['bureau']
prev_app = dfs['prev_app']

print("=== 6. PHÂN TÍCH LỊCH SỬ TÍN DỤNG (BUREAU & PREVIOUS APP) ===\n")

# --- BƯỚC 1: GOM NHÓM DỮ LIỆU (AGGREGATION) ---

# 1.1. Xử lý bảng Bureau (Lịch sử ở nơi khác)
# Tính: Tổng số khoản vay, Số khoản đang còn nợ (Active)
grp_bureau = bureau.groupby('SK_ID_CURR').agg({
    'SK_ID_BUREAU': 'count',                 # Đếm số lượng khoản vay
    'CREDIT_ACTIVE': lambda x: (x == 'Active').sum() # Đếm số khoản đang Active
}).reset_index()
grp_bureau.rename(columns={'SK_ID_BUREAU': 'BUREAU_COUNT', 
                           'CREDIT_ACTIVE': 'BUREAU_ACTIVE_COUNT'}, inplace=True)

# 1.2. Xử lý bảng Previous Application (Lịch sử tại Home Credit)
# Tính: Tổng số đơn vay, Số lần bị từ chối
grp_prev = prev_app.groupby('SK_ID_CURR').agg({
    'SK_ID_PREV': 'count',
    'NAME_CONTRACT_STATUS': lambda x: (x == 'Refused').sum()
}).reset_index()
grp_prev.rename(columns={'SK_ID_PREV': 'PREV_APP_COUNT', 
                         'NAME_CONTRACT_STATUS': 'PREV_REFUSED_COUNT'}, inplace=True)

# --- BƯỚC 2: MERGE VÀO BẢNG CHÍNH ---
df_merge = df_train.merge(grp_bureau, on='SK_ID_CURR', how='left')
df_merge = df_merge.merge(grp_prev, on='SK_ID_CURR', how='left')

# Điền 0 cho những người không có lịch sử (NaN -> 0)
df_merge.fillna(0, inplace=True)

# --- BƯỚC 3: THỐNG KÊ MÔ TẢ SAU KHI GOM ---
print("Thống kê trung bình:")
print(df_merge.groupby('TARGET')[['BUREAU_COUNT', 'PREV_REFUSED_COUNT']].mean())


# --- BƯỚC 4: VẼ BIỂU ĐỒ ---
fig = plt.figure(figsize=(16, 6))

# Biểu đồ 1: Số lượng khoản vay quá khứ (Bureau) ảnh hưởng thế nào?
# Dùng Boxplot để so sánh phân phối, loại bỏ outlier quá lớn để dễ nhìn (showfliers=False)
ax1 = fig.add_subplot(1, 2, 1)
sns.boxplot(x='TARGET', y='BUREAU_COUNT', data=df_merge, palette='Set3', showfliers=False, ax=ax1)
ax1.set_title('Số lượng khoản vay ở tổ chức khác (Bureau Count)')
ax1.set_xlabel('Target (0: Tốt, 1: Xấu)')
ax1.set_ylabel('Số lượng khoản vay')

# Biểu đồ 2: Số lần bị từ chối trong quá khứ (Refused Count)
# Đây là biến cực mạnh
ax2 = fig.add_subplot(1, 2, 2)
# Tạo bảng phụ tính tỷ lệ vỡ nợ theo số lần từ chối
refused_risk = df_merge.groupby('PREV_REFUSED_COUNT')['TARGET'].mean()
# Chỉ lấy đến 5 lần từ chối (vì số liệu >5 rất ít, gây nhiễu)
refused_risk = refused_risk[refused_risk.index <= 5]

sns.barplot(x=refused_risk.index, y=refused_risk.values, palette='Reds', ax=ax2)
ax2.set_title('Tỷ lệ Vỡ nợ theo Số lần Bị từ chối trước đây')
ax2.set_xlabel('Số lần bị từ chối (Refused)')
ax2.set_ylabel('Tỷ lệ Vỡ nợ trung bình')

# Gắn số
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.1%}', (p.get_x() + 0.25, p.get_height() + 0.002))

plt.tight_layout()
plt.show()
```

### 7. Phân tích Hành vi thanh toán

```python
# Lấy dữ liệu
ins = dfs['installments'].copy()
df_train = dfs['app_train'][['SK_ID_CURR', 'TARGET']].copy()

print("=== 7. PHÂN TÍCH HÀNH VI THANH TOÁN (INSTALLMENTS) ===\n")

# --- BƯỚC 1: FEATURE ENGINEERING (Tạo biến hành vi) ---

# 1.1. Tính số ngày trả chậm (Days Past Due - DPD)
# Nếu (Ngày thực trả - Ngày phải trả) > 0 là Chậm, ngược lại là 0 (Sớm/Đúng hạn)
ins['DPD'] = ins['DAYS_ENTRY_PAYMENT'] - ins['DAYS_INSTALMENT']
ins['DPD'] = ins['DPD'].apply(lambda x: x if x > 0 else 0)

# 1.2. Tính số tiền trả thiếu (Underpayment)
# Nếu (Phải trả - Thực trả) > 0 là Thiếu
ins['PAYMENT_DIFF'] = ins['AMT_INSTALMENT'] - ins['AMT_PAYMENT']
ins['PAYMENT_DIFF'] = ins['PAYMENT_DIFF'].apply(lambda x: x if x > 0 else 0)

# --- BƯỚC 2: GOM NHÓM THEO KHÁCH HÀNG ---
# Ta quan tâm: Khách hàng này trung bình trả chậm bao nhiêu ngày? Có bao giờ trả thiếu không?
ins_agg = ins.groupby('SK_ID_CURR').agg({
    'DPD': ['mean', 'max', lambda x: (x > 0).mean()], # Trung bình ngày chậm, Max ngày chậm, Tỷ lệ lần trả chậm
    'PAYMENT_DIFF': ['mean', lambda x: (x > 0).sum()] # Trung bình tiền thiếu, Số lần trả thiếu
})

# Đổi tên cột cho gọn (Flatten MultiIndex)
ins_agg.columns = ['AVG_DPD', 'MAX_DPD', 'LATE_PAYMENT_RATIO', 'AVG_UNDERPAID', 'UNDERPAID_COUNT']
ins_agg.reset_index(inplace=True)

# Merge vào bảng train
df_pay = df_train.merge(ins_agg, on='SK_ID_CURR', how='left')
df_pay.fillna(0, inplace=True) # Ai không có lịch sử coi như là 0


# --- BƯỚC 3: VẼ BIỂU ĐỒ ---
fig = plt.figure(figsize=(16, 6))

# Biểu đồ 1: So sánh Tỷ lệ vỡ nợ giữa người "Chưa từng trả chậm" và "Từng trả chậm"
# Tạo cột phân loại tạm thời
df_pay['HAS_BEEN_LATE'] = df_pay['MAX_DPD'].apply(lambda x: 'Từng trễ hạn' if x > 0 else 'Luôn đúng hạn')

ax1 = fig.add_subplot(1, 2, 1)
sns.barplot(x='HAS_BEEN_LATE', y='TARGET', data=df_pay, palette='Paired', ax=ax1)
ax1.set_title('Tác động của Lịch sử Trễ hạn (Late Payment)')
ax1.set_ylabel('Tỷ lệ Vỡ nợ trung bình')
# Gắn số
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.1%}', (p.get_x() + 0.3, p.get_height() + 0.002))

# Biểu đồ 2: Phân phối số ngày trả chậm trung bình (Log scale vì chênh lệch lớn)
# Chỉ xét những người Target=1 để xem họ thường chậm bao lâu
ax2 = fig.add_subplot(1, 2, 2)
# Lọc bỏ những người luôn đúng hạn để biểu đồ đỡ bị dồn về 0
data_late = df_pay[df_pay['AVG_DPD'] > 0] 

sns.boxplot(x='TARGET', y='AVG_DPD', data=data_late, palette='Set1', ax=ax2)
ax2.set_yscale('log') # Dùng Log scale trục Y
ax2.set_title('Phân phối Số ngày quá hạn trung bình (Log Scale)')
ax2.set_ylabel('Số ngày quá hạn (AVG_DPD)')
ax2.set_xlabel('Target (0: Tốt, 1: Xấu)')

plt.tight_layout()
plt.show()

# In thống kê cụ thể
print("Thống kê trung bình (Hành vi trả nợ):")
print(df_pay.groupby('TARGET')[['AVG_DPD', 'LATE_PAYMENT_RATIO', 'AVG_UNDERPAID']].mean())
```

### 9. Phát hiện giá trị ngoại lai

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Lấy dữ liệu
df = dfs['app_train'].copy()

print("=== 9. PHÂN TÍCH TOÀN DIỆN GIÁ TRỊ NGOẠI LAI (OUTLIERS) ===\n")

# --- DANH SÁCH CÁC BIẾN QUAN TRỌNG CẦN KIỂM TRA ---
# Bao gồm: Lỗi dữ liệu, Tài chính, Số đếm, Dân số
cols_to_check = [
    'DAYS_EMPLOYED',                # Số ngày làm việc (Lỗi 365243)
    'AMT_INCOME_TOTAL',             # Thu nhập (Cực đoan)
    'CNT_CHILDREN',                 # Số con (Hiếm)
    'AMT_CREDIT',                   # Số tiền vay (Tài chính lệch phải)
    'AMT_ANNUITY',                  # Số tiền trả góp (Tài chính lệch phải)
    'AMT_GOODS_PRICE',              # Giá trị hàng hóa
    'OBS_30_CNT_SOCIAL_CIRCLE',     # Mạng lưới xã hội (Giá trị lạ)
    'REGION_POPULATION_RELATIVE'    # Mật độ dân số (Có thành phố cực lớn)
]

# --- 1. BẢNG THỐNG KÊ CHI TIẾT (QUÉT DIỆN RỘNG) ---
outlier_stats = []

for col in cols_to_check:
    series = df[col].dropna()
    
    # Tính các mốc thống kê
    Min = series.min()
    Max = series.max()
    Mean = series.mean()
    Median = series.median()
    
    # Tính IQR để xác định outlier theo thống kê
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    
    # Đặc biệt với DAYS_EMPLOYED: Outlier là số dương lớn
    if col == 'DAYS_EMPLOYED':
        n_outliers = (series == 365243).sum()
        rule = "Giá trị == 365243"
    else:
        # Với các biến tài chính/số đếm, ta quan tâm đuôi phải (giá trị quá lớn)
        n_outliers = (series > upper_bound).sum()
        rule = f"> {upper_bound:,.2f}"

    pct = (n_outliers / len(series)) * 100
    
    outlier_stats.append({
        'Feature': col,
        'Min': Min,
        'Median': Median,
        'Max': Max,
        'Outliers Count': n_outliers,
        'Outliers (%)': round(pct, 2),
        'Detection Rule': rule
    })

# Hiển thị bảng kết quả
outlier_df = pd.DataFrame(outlier_stats)
print("Bảng Tổng hợp Tình trạng Ngoại lai:")
display(outlier_df.style.background_gradient(cmap='Reds', subset=['Outliers (%)']))


# --- 2. TRỰC QUAN HÓA (CHỈ CÁC BIẾN QUAN TRỌNG NHẤT) ---
# Chọn 6 biến đặc trưng nhất để vẽ
fig = plt.figure(figsize=(18, 10))
plt.suptitle('Trực quan hóa các loại Ngoại lai khác nhau', fontsize=16)

# 2.1. DAYS_EMPLOYED (Histogram để thấy sự tách biệt)
ax1 = fig.add_subplot(2, 3, 1)
sns.histplot(df['DAYS_EMPLOYED'], bins=50, color='red', ax=ax1)
ax1.set_title('DAYS_EMPLOYED: Lỗi dữ liệu (Artifact)')
ax1.set_xlabel('Ngày làm việc (Dương là lỗi)')

# 2.2. AMT_INCOME_TOTAL (Boxplot)
ax2 = fig.add_subplot(2, 3, 2)
sns.boxplot(x=df['AMT_INCOME_TOTAL'], color='green', ax=ax2)
ax2.set_title('AMT_INCOME_TOTAL: Giá trị Cực đoan (Extreme)')
ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# 2.3. CNT_CHILDREN (Countplot)
ax3 = fig.add_subplot(2, 3, 3)
sns.countplot(x='CNT_CHILDREN', data=df, palette='viridis', ax=ax3)
ax3.set_title('CNT_CHILDREN: Trường hợp Hiếm (Rare)')
ax3.set_yscale('log') # Log scale để nhìn thấy cột 19 con

# 2.4. AMT_CREDIT (Boxplot - Tài chính thông thường)
ax4 = fig.add_subplot(2, 3, 4)
sns.boxplot(x=df['AMT_CREDIT'], color='skyblue', ax=ax4)
ax4.set_title('AMT_CREDIT: Phân phối lệch phải (Skewed)')
ax4.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

# 2.5. OBS_30_CNT_SOCIAL_CIRCLE (Scatter/Boxplot)
ax5 = fig.add_subplot(2, 3, 5)
sns.boxplot(x=df['OBS_30_CNT_SOCIAL_CIRCLE'], color='orange', ax=ax5)
ax5.set_title('OBS_30_SOCIAL: Mạng lưới xã hội bất thường')

# 2.6. REGION_POPULATION_RELATIVE (Distribution)
ax6 = fig.add_subplot(2, 3, 6)
sns.histplot(df['REGION_POPULATION_RELATIVE'], bins=50, color='purple', ax=ax6)
ax6.set_title('REGION_POPULATION: Thành phố lớn tách biệt')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# CHUẨN BỊ: Load lại dữ liệu gốc từ dictionary
import numpy as np
import pandas as pd
import gc

df_train = dfs['app_train'].copy()
df_test = dfs['app_test'].copy()

print(f"Kích thước ban đầu - Train: {df_train.shape}, Test: {df_test.shape}")
```

a. Xử lý lỗi đặc biệt

```python
# --- 1: XỬ LÝ LỖI & NGÀY THÁNG ---

# 1. Xử lý lỗi 365243 (1000 năm) ở cột DAYS_EMPLOYED
# Áp dụng cho cả Train và Test
df_train['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
df_test['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)

# 2. Chuyển tất cả cột DAYS_... thành số dương
# Tìm các cột bắt đầu bằng 'DAYS_'
date_cols = [col for col in df_train.columns if col.startswith('DAYS_')]

for col in date_cols:
    df_train[col] = df_train[col].abs()
    df_test[col] = df_test[col].abs()

# 3. Tạo các cột tính theo "Năm" (Feature Engineering cơ bản)
for df in [df_train, df_test]:
    df['AGE'] = df['DAYS_BIRTH'] / 365
    df['YEARS_EMPLOYED'] = df['DAYS_EMPLOYED'] / 365
    df['YEARS_REGISTRATION'] = df['DAYS_REGISTRATION'] / 365
    df['YEARS_ID_PUBLISH'] = df['DAYS_ID_PUBLISH'] / 365

print("Đã xử lý lỗi 365243 và tạo 4 cột mới")
```

b. Xử lý giá trị ngoại lai

```python
# --- 2: XỬ LÝ NGOẠI LAI (OUTLIERS) ---

# 1. Xử lý Số con (CNT_CHILDREN)
# Ai có trên 5 con thì quy về 5 (Áp dụng cả Train và Test để đồng bộ)
df_train.loc[df_train['CNT_CHILDREN'] > 5, 'CNT_CHILDREN'] = 5
df_test.loc[df_test['CNT_CHILDREN'] > 5, 'CNT_CHILDREN'] = 5

# 2. Xử lý Thu nhập (AMT_INCOME_TOTAL) - CHỈ LÀM TRÊN TRAIN
# Tìm người có thu nhập 117 Triệu (hoặc > 100M) để xóa
print(f"Số dòng Train trước khi xóa outlier: {len(df_train)}")
df_train = df_train[df_train['AMT_INCOME_TOTAL'] < 100000000]
print(f"Số dòng Train sau khi xóa outlier : {len(df_train)}")

# Tập Test KHÔNG xóa dòng nào cả.

print("Đã chặn trên số con và xóa người có thu nhập 117 triệu khỏi tập Train.")
```

c. Biến đổi dặc trưng

```python
# --- 3: TẠO ĐẶC TRƯNG TÀI CHÍNH MỚI ---

# Viết vòng lặp để áp dụng công thức cho cả 2 bảng
for df in [df_train, df_test]:
    # Tỷ lệ Vay / Thu nhập (Vay gấp mấy lần lương)
    df['CREDIT_TO_INCOME'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
    
    # Tỷ lệ Trả góp / Thu nhập (Gánh nặng nợ hàng tháng)
    df['ANNUITY_TO_INCOME'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
    
    # Thời hạn vay giả định (Tháng) = Tiền vay / Tiền trả mỗi kỳ
    df['CREDIT_TERM'] = df['AMT_ANNUITY'] / df['AMT_CREDIT']
    
    # Tỷ lệ Giá hàng / Tiền vay (Xem có vay lố hay không)
    df['GOODS_TO_CREDIT'] = df['AMT_GOODS_PRICE'] / df['AMT_CREDIT']

print("Đã tạo 4 biến tỷ lệ tài chính")
```

### **GIAI ĐOẠN 2: XỬ LÝ BẢNG PHỤ**

```python
import numpy as np
import pandas as pd
import gc

# Hàm One-Hot Encoding (Chuyển chữ thành số)
def one_hot_encoder(df, nan_as_category=True):
    original_columns = list(df.columns)
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    
    # Tạo One-Hot (dummy variables)
    df = pd.get_dummies(df, columns=categorical_columns, dummy_na=nan_as_category)
    
    # Lấy danh sách tên các cột mới tạo ra
    new_columns = [c for c in df.columns if c not in original_columns]
    return df, new_columns

print("Đã khởi tạo hàm one_hot_encoder.")
```

a. Xử lý nhóm tín dụng bên ngoài

```python
print("--- 1. ĐANG XỬ LÝ BẢNG BUREAU & BALANCE ---")

# 1. Load dữ liệu (Sử dụng key rút gọn: bureau_bal)
bureau = dfs['bureau'].copy()
bb = dfs['bureau_bal'].copy()  # <--- Đã sửa chỗ này

# 2. Xử lý bảng Bureau Balance trước (One-Hot & Aggregate)
bb, bb_cat = one_hot_encoder(bb, nan_as_category=True)
# Gom nhóm theo SK_ID_BUREAU
bb_agg = bb.groupby('SK_ID_BUREAU').agg({
    'MONTHS_BALANCE': ['min', 'max', 'size']
})
# Gom nhóm các cột One-Hot
for col in bb_cat:
    bb_agg[col] = bb.groupby('SK_ID_BUREAU')[col].agg('mean')

# Làm phẳng tên cột
bb_agg.columns = [f'BB_{c[0]}_{c[1].upper()}' for c in bb_agg.columns.tolist()]
bb = bb_agg

# 3. Nối kết quả Balance vào bảng Bureau
bureau = bureau.join(bb, how='left', on='SK_ID_BUREAU')
bureau.drop(columns=['SK_ID_BUREAU'], inplace=True)

# 4. Xử lý bảng Bureau
# Thay thế lỗi 365243
cols_days_bureau = ['DAYS_CREDIT', 'DAYS_CREDIT_ENDDATE', 'DAYS_ENDDATE_FACT', 'DAYS_CREDIT_UPDATE']
for col in cols_days_bureau:
    if col in bureau.columns:
        bureau[col].replace(365243, np.nan, inplace=True)

# One-Hot Encoding
bureau, bureau_cat = one_hot_encoder(bureau, nan_as_category=True)

# Gom nhóm theo SK_ID_CURR
num_aggregations = {
    'DAYS_CREDIT': ['min', 'max', 'mean', 'var'],
    'DAYS_CREDIT_ENDDATE': ['min', 'max', 'mean'],
    'DAYS_CREDIT_UPDATE': ['mean'],
    'CREDIT_DAY_OVERDUE': ['max', 'mean'],
    'AMT_CREDIT_MAX_OVERDUE': ['mean'],
    'AMT_CREDIT_SUM': ['max', 'mean', 'sum'],
    'AMT_CREDIT_SUM_DEBT': ['max', 'mean', 'sum'],
    'AMT_CREDIT_SUM_OVERDUE': ['mean'],
    'CNT_CREDIT_PROLONG': ['sum'],
    'BB_MONTHS_BALANCE_MIN': ['min'],
    'BB_MONTHS_BALANCE_MAX': ['max'],
    'BB_MONTHS_BALANCE_SIZE': ['mean', 'sum']
}
cat_aggregations = {}
for cat in bureau_cat: cat_aggregations[cat] = ['mean']

bureau_agg = bureau.groupby('SK_ID_CURR').agg({**num_aggregations, **cat_aggregations})
bureau_agg.columns = pd.Index([f'BURO_{e[0]}_{e[1].upper()}' for e in bureau_agg.columns.tolist()])

print(f"-> Hoàn tất Bureau. Kích thước: {bureau_agg.shape}")
gc.collect()
```

b. XỬ LÝ BẢNG PREVIOUS APPLICATION (Đơn vay cũ)

```python
print("--- 2. ĐANG XỬ LÝ PREVIOUS APPLICATION ---")

# Sử dụng key rút gọn: prev_app
prev = dfs['prev_app'].copy()

# Xử lý lỗi 365243
days_cols_prev = ['DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE', 'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION']
for col in days_cols_prev:
    prev[col].replace(365243, np.nan, inplace=True)

# Feature Engineering
prev['APP_CREDIT_PERC'] = prev['AMT_APPLICATION'] / prev['AMT_CREDIT']

# One-Hot Encoding
prev, cat_cols = one_hot_encoder(prev, nan_as_category=True)

# Gom nhóm
num_aggregations = {
    'AMT_ANNUITY': ['min', 'max', 'mean'],
    'AMT_APPLICATION': ['min', 'max', 'mean'],
    'AMT_CREDIT': ['min', 'max', 'mean'],
    'APP_CREDIT_PERC': ['min', 'max', 'mean', 'var'],
    'AMT_DOWN_PAYMENT': ['min', 'max', 'mean'],
    'AMT_GOODS_PRICE': ['min', 'max', 'mean'],
    'HOUR_APPR_PROCESS_START': ['min', 'max', 'mean'],
    'RATE_DOWN_PAYMENT': ['min', 'max', 'mean'],
    'DAYS_DECISION': ['min', 'max', 'mean'],
    'CNT_PAYMENT': ['mean', 'sum'],
}
cat_aggregations = {}
for cat in cat_cols: cat_aggregations[cat] = ['mean']

prev_agg = prev.groupby('SK_ID_CURR').agg({**num_aggregations, **cat_aggregations})
prev_agg.columns = pd.Index([f'PREV_{e[0]}_{e[1].upper()}' for e in prev_agg.columns.tolist()])

print(f"-> Hoàn tất Prev App. Kích thước: {prev_agg.shape}")
gc.collect()
```

c. XỬ LÝ BẢNG POS CASH & INSTALLMENTS (Hành vi thanh toán)

```python
print("--- 3. ĐANG XỬ LÝ POS CASH & INSTALLMENTS ---")

# --- A. POS CASH BALANCE ---
pos = dfs['pos_cash'].copy()
pos, cat_cols = one_hot_encoder(pos, nan_as_category=True)

# Gom nhóm (Tính toán DPD là quan trọng nhất)
aggregations = {
    'MONTHS_BALANCE': ['max', 'mean', 'size'],
    'SK_DPD': ['max', 'mean'],
    'SK_DPD_DEF': ['max', 'mean']
}
for cat in cat_cols: aggregations[cat] = ['mean']

pos_agg = pos.groupby('SK_ID_CURR').agg(aggregations)
pos_agg.columns = pd.Index([f'POS_{e[0]}_{e[1].upper()}' for e in pos_agg.columns.tolist()])


# --- B. INSTALLMENTS PAYMENTS ---
ins = dfs['installments'].copy()

# Feature Engineering
# 1. Trả chậm bao nhiêu ngày? (Chỉ lấy số dương)
ins['DPD'] = ins['DAYS_ENTRY_PAYMENT'] - ins['DAYS_INSTALMENT']
ins['DPD'] = ins['DPD'].apply(lambda x: x if x > 0 else 0)

# 2. Trả thiếu bao nhiêu tiền? (Chỉ lấy số dương)
ins['PAYMENT_DIFF'] = ins['AMT_INSTALMENT'] - ins['AMT_PAYMENT']
ins['PAYMENT_DIFF'] = ins['PAYMENT_DIFF'].apply(lambda x: x if x > 0 else 0)

# Gom nhóm
aggregations = {
    'NUM_INSTALMENT_VERSION': ['nunique'],
    'DPD': ['max', 'mean', 'sum', 'var'],          # Quan trọng
    'PAYMENT_DIFF': ['max', 'mean', 'sum', 'var'], # Quan trọng
    'AMT_INSTALMENT': ['max', 'mean', 'sum'],
    'AMT_PAYMENT': ['min', 'max', 'mean', 'sum'],
    'DAYS_ENTRY_PAYMENT': ['max', 'mean', 'sum']
}

ins_agg = ins.groupby('SK_ID_CURR').agg(aggregations)
ins_agg.columns = pd.Index([f'INSTAL_{e[0]}_{e[1].upper()}' for e in ins_agg.columns.tolist()])

print(f"-> Hoàn tất POS ({pos_agg.shape}) & Installments ({ins_agg.shape})")
gc.collect()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
import gc

print("=== GIAI ĐOẠN 5: CHUẨN BỊ DỮ LIỆU CUỐI CÙNG (FINAL PREP) ===\n")

# =============================================================================
# BƯỚC 0: XỬ LÝ GIÁ TRỊ VÔ CỰC (INF)
# =============================================================================
# Khi chia cho 0 (ví dụ tạo các tỷ lệ tài chính), ta có thể tạo ra np.inf.
# Cần chuyển inf về NaN để Imputer xử lý.
print("0. Đang kiểm tra và xử lý giá trị vô cực (inf)...")
df_train.replace([np.inf, -np.inf], np.nan, inplace=True)
df_test.replace([np.inf, -np.inf], np.nan, inplace=True)


# =============================================================================
# BƯỚC 1: XỬ LÝ MISSING VALUE (ĐIỀN GIÁ TRỊ THIẾU)
# =============================================================================
print("\n1. Đang xử lý Missing Value (Chiến lược: Điền bằng Median)...")

# Lấy danh sách các cột số (trừ ID và TARGET)
# Lưu ý: Không điền missing cho cột object (chữ), ta sẽ để One-Hot xử lý
numeric_feats = [col for col in df_train.columns if df_train[col].dtype != 'object' and col not in ['SK_ID_CURR', 'TARGET']]

# Sử dụng SimpleImputer để điền Median (nhanh và ít bị ảnh hưởng bởi outlier)
imputer = SimpleImputer(strategy='median')

# Fit trên Train, Transform trên cả Train và Test để tránh Data Leakage
imputer.fit(df_train[numeric_feats])

df_train[numeric_feats] = imputer.transform(df_train[numeric_feats])
df_test[numeric_feats] = imputer.transform(df_test[numeric_feats])

print(f"-> Đã điền xong giá trị thiếu cho {len(numeric_feats)} cột số.")


# =============================================================================
# BƯỚC 2: MÃ HÓA BẢNG CHÍNH (ONE-HOT ENCODING)
# =============================================================================
print("\n2. Đang mã hóa các cột phân loại (One-Hot Encoding)...")

# Lấy các cột là object (chữ) còn lại trong bảng chính
cat_cols = [col for col in df_train.columns if df_train[col].dtype == 'object']
print(f"-> Các cột cần mã hóa: {cat_cols}")

# Thực hiện One-Hot
df_train = pd.get_dummies(df_train, columns=cat_cols)
df_test = pd.get_dummies(df_test, columns=cat_cols)

# CỰC KỲ QUAN TRỌNG: Đồng bộ cột giữa Train và Test
# (Ví dụ: Train có giới tính 'XNA' mà Test không có -> lệch cột -> lỗi model)
train_labels = df_train['TARGET'] # Giữ lại Target
train_ids = df_train['SK_ID_CURR']
test_ids = df_test['SK_ID_CURR']

# Căn chỉnh 2 bảng (chỉ giữ lại các cột xuất hiện ở cả 2)
df_train, df_test = df_train.align(df_test, join='inner', axis=1)

# Trả lại các cột quan trọng
df_train['TARGET'] = train_labels
# (SK_ID_CURR đã có trong inner join rồi)

print(f"-> Sau khi mã hóa và đồng bộ: Train {df_train.shape}, Test {df_test.shape}")


# =============================================================================
# BƯỚC 3: CHUẨN HÓA DỮ LIỆU (SCALING [0, 1])
# =============================================================================
print("\n3. Đang chuẩn hóa dữ liệu (MinMaxScaler về [0, 1])...")

# Danh sách các cột cần Scale (Tất cả các cột Feature, trừ ID và Target)
features_to_scale = [col for col in df_train.columns if col not in ['SK_ID_CURR', 'TARGET']]

scaler = MinMaxScaler(feature_range=(0, 1))

# Fit trên Train
scaler.fit(df_train[features_to_scale])

# Transform trên cả hai
df_train[features_to_scale] = scaler.transform(df_train[features_to_scale])
df_test[features_to_scale] = scaler.transform(df_test[features_to_scale])

print("-> Đã hoàn tất chuẩn hóa dữ liệu.")
print("-" * 40)

# Xem trước dữ liệu sau khi xử lý
print("Dữ liệu Train sau khi Final Prep (5 dòng đầu):")
display(df_train.head())

# Dọn dẹp bộ nhớ
gc.collect()
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
import re
from sklearn.metrics import (accuracy_score, classification_report, 
                             roc_auc_score, roc_curve, confusion_matrix)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

print("=== CHUYỂN SANG LIGHTGBM (GRADIENT BOOSTING) ===\n")

# 1. Sửa tên cột (Bắt buộc với LightGBM)
# Loại bỏ các ký tự đặc biệt trong tên cột để tránh lỗi JSON
def clean_col_names(df):
    df = df.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
    return df

X_train = clean_col_names(X_train)
X_val = clean_col_names(X_val)

# 2. Khởi tạo mô hình
# Đây là bộ tham số (Hyperparams) khá chuẩn cho Home Credit
lgbm_model = LGBMClassifier(
    n_estimators=2000,        # Số lượng cây tối đa
    learning_rate=0.02,       # Tốc độ học chậm để dò tìm điểm tối ưu kỹ hơn
    num_leaves=34,            # Độ phức tạp của cây (thường < 2^max_depth)
    max_depth=8,              # Độ sâu tối đa
    colsample_bytree=0.9,     # Mỗi cây chỉ dùng 90% số cột (tránh overfit)
    subsample=0.8,            # Mỗi cây chỉ dùng 80% số dòng
    reg_alpha=0.04,           # L1 Regularization
    reg_lambda=0.07,          # L2 Regularization
    class_weight='balanced',  # <--- Tự động cân bằng mẫu (92% vs 8%)
    random_state=42,
    n_jobs=-1
)

# 3. Huấn luyện với Early Stopping
# Nếu sau 100 cây mà kết quả không tốt lên thì dừng lại để tiết kiệm thời gian
print("--- Đang huấn luyện (Sẽ nhanh hơn Random Forest)... ---")
lgbm_model.fit(
    X_train, y_train,
    eval_set=[(X_train, y_train), (X_val, y_val)],
    eval_metric='auc',
    callbacks=[lgb.early_stopping(stopping_rounds=100), lgb.log_evaluation(200)]
)

print("\n-> Huấn luyện hoàn tất!")
```

```python
# 1. Dự báo
y_pred_lgbm = lgbm_model.predict(X_val)            # Nhãn (0/1)
y_pred_proba_lgbm = lgbm_model.predict_proba(X_val)[:, 1] # Xác suất

# 2. In báo cáo
print("\n=== BÁO CÁO HIỆU SUẤT LIGHTGBM ===")
print(f"1. Accuracy : {accuracy_score(y_val, y_pred_lgbm):.4f}")
print(f"2. ROC-AUC  : {roc_auc_score(y_val, y_pred_proba_lgbm):.4f}")
print("-" * 50)
print(classification_report(y_val, y_pred_lgbm))

# 3. Vẽ biểu đồ
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# Confusion Matrix
cm = confusion_matrix(y_val, y_pred_lgbm)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=False, ax=ax[0], annot_kws={"size": 14})
ax[0].set_title('Confusion Matrix (LightGBM)', fontsize=14)
ax[0].set_xlabel('Dự báo')
ax[0].set_ylabel('Thực tế')

# ROC Curve
fpr, tpr, _ = roc_curve(y_val, y_pred_proba_lgbm)
auc_score = roc_auc_score(y_val, y_pred_proba_lgbm)
ax[1].plot(fpr, tpr, color='green', lw=2, label=f'LightGBM (AUC = {auc_score:.3f})')
ax[1].plot([0, 1], [0, 1], color='navy', linestyle='--')
ax[1].set_title('ROC Curve', fontsize=14)
ax[1].legend(loc="lower right")
ax[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
